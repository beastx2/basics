```python
df4 = df[(df['release_year'] >=2000) & (df['release_year'] <= 2005)]
```

```python
df4.info()
```

```python
df4 = df4.sort_values(by=['release_year'])
df4
```

```python
di = df4['original_title'].str.split().apply(len).value_counts()
di
```

```python
total_count = 0
for index,value in di.items():
    total_count += index*value
total_count
```

```python
no_of_movies = len(df4.index)
avg_no_of_words = total_count/no_of_movies
round(avg_no_of_words)
```
